## TF-IDF (Term Frequency-Inverse Document Frequency)

**TF-IDF** (Term Frequency-Inverse Document Frequency) adalah suatu metode yang digunakan dalam pemrosesan bahasa alami (NLP) dan informasi retrieval untuk menilai seberapa penting suatu kata dalam suatu dokumen relatif terhadap koleksi atau kumpulan dokumen (corpus). TF-IDF sering digunakan untuk analisis teks, pengambilan informasi, dan pembuatan sistem rekomendasi.

**Komponen TF-IDF**

TF-IDF terdiri dari dua komponen utama:

1. **Term Frequency (TF)**:
   - Mengukur seberapa sering suatu kata muncul dalam sebuah dokumen. Semakin sering kata muncul, semakin besar bobotnya.
   - Rumus untuk menghitung TF dari suatu kata $( t )$ dalam dokumen $( d )$:
   $$
   \text{TF}(t, d) = \frac{\text{Jumlah kemunculan kata } t \text{ dalam } d}{\text{Total jumlah kata dalam } d}
   $$

2. **Inverse Document Frequency (IDF)**:
   - Mengukur seberapa penting kata tersebut dalam seluruh koleksi dokumen. Kata yang muncul di banyak dokumen memiliki bobot rendah, sedangkan kata yang muncul di sedikit dokumen memiliki bobot lebih tinggi.
   - Rumus untuk menghitung IDF dari suatu kata $( t )$:
   $$
   \text{IDF}(t) = \log\left(\frac{\text{Total jumlah dokumen}}{\text{Jumlah dokumen yang mengandung kata } t}\right)
   $$
   - Dalam praktiknya, sering ditambahkan satu untuk menghindari pembagian dengan nol:
   $$
   \text{IDF}(t) = \log\left(\frac{N}{\text{df}(t)} + 1\right)
   $$
   - Di mana \( N \) adalah jumlah total dokumen dan $( \text{df}(t) )$ adalah jumlah dokumen yang mengandung kata $( t )$.

**Menghitung TF-IDF**

Setelah menghitung TF dan IDF, nilai TF-IDF untuk suatu kata $( t )$ dalam dokumen $( d )$ dapat dihitung dengan rumus:
$$
\text{TF-IDF}(t, d) = \text{TF}(t, d) \times \text{IDF}(t)
$$

**Interpretasi**

- Nilai Tinggi: Jika nilai TF-IDF tinggi, ini menunjukkan bahwa kata tersebut sering muncul dalam dokumen tertentu tetapi jarang muncul di seluruh koleksi dokumen. Kata ini dianggap penting untuk dokumen tersebut.
- Nilai Rendah: Jika nilai TF-IDF rendah, ini bisa berarti kata tersebut umum dan muncul di banyak dokumen, sehingga tidak memberikan banyak informasi khusus tentang dokumen tersebut.


## Cosine Similarity

Cosine similarity adalah ukuran yang digunakan untuk menentukan seberapa mirip dua vektor dalam ruang multidimensi berdasarkan sudut di antara keduanya. Dalam konteks pemrosesan bahasa alami, cosine similarity sering digunakan untuk mengukur kesamaan antara teks yang direpresentasikan sebagai vektor, biasanya menggunakan teknik seperti TF-IDF atau word embeddings.

**Konsep Dasar Cosine Similarity**

Cosine similarity dihitung dengan menggunakan rumus berikut:

$$
\text{Cosine Similarity} = \frac{A \cdot B}{\|A\| \|B\|}
$$

Di mana:
- $(A)$ dan $(B)$ adalah dua vektor yang akan dibandingkan.
- $(A \cdot B)$ adalah hasil kali dot (dot product) dari kedua vektor, yang mengukur seberapa besar dua vektor tersebut saling berhubungan.
- $(|A|)$ dan $(|B|)$ adalah norma (magnitude) dari vektor $(A)$ dan $(B)$, yang menunjukkan panjang masing-masing vektor.

**Nilai Cosine Similarity**

Rentang Nilai:

1: Menunjukkan dua dokumen sangat mirip (atau identik).
0: Menunjukkan tidak ada kesamaan (dokumen tidak mirip).
Dalam konteks pengolahan teks, nilai cosine similarity umumnya dianggap dalam rentang 0 hingga 1. Ini karena vektor representasi teks (seperti TF-IDF atau word embeddings) biasanya memiliki nilai positif.

Interpretasi:

Nilai 1: Dokumen A dan B identik dalam konteks konten.
Nilai 0.5: Menunjukkan kesamaan moderat, artinya dokumen memiliki beberapa kata atau konsep yang sama.
Nilai 0: Dokumen tidak memiliki kesamaan atau tidak berbagi kata kunci.

## Matriks Adjacency

Matriks adjacency adalah representasi matematis dari graf yang menunjukkan hubungan antara node (simpul) dalam graf tersebut. Dalam konteks pemrosesan bahasa alami dan analisis graph, matriks adjacency sering digunakan untuk merepresentasikan hubungan antar elemen, seperti kalimat dalam sebuah dokumen, berdasarkan kriteria tertentu.

**Konsep Dasar Matriks Adjacency**

1. Definisi: 
   Matriks adjacency adalah matriks persegi yang digunakan untuk merepresentasikan graf. Setiap elemen dalam matriks menunjukkan ada tidaknya hubungan antara dua node. Dalam matriks adjacency, elemen pada baris ke-$(i)$ dan kolom ke-$(j)$ akan bernilai:
   - **1** (atau nilai positif) jika ada edge (hubungan) antara node $(i)$ dan node $(j)$.
   - **0** jika tidak ada edge antara keduanya.

2. Ukuran Matriks: 
   Matriks adjacency untuk graf dengan $(n)$ node memiliki ukuran $(n \times n)$, di mana $(n)$ adalah jumlah total node dalam graf. Baris dan kolom pada matriks mewakili node yang ada dalam graf.

**Contoh Matriks Adjacency**

Misalkan kita memiliki graf sederhana dengan 4 node (A, B, C, dan D) dan hubungan sebagai berikut:
- A terhubung ke B dan C.
- B terhubung ke C.
- C terhubung ke D.

Matriks adjacency untuk graf ini dapat dituliskan sebagai berikut:

$$
\begin{matrix}
  & A & B & C & D \\
A & 0 & 1 & 1 & 0 \\
B & 0 & 0 & 1 & 0 \\
C & 0 & 0 & 0 & 1 \\
D & 0 & 0 & 0 & 0 \\
\end{matrix}
$$

Dalam matriks di atas:
- Baris A menunjukkan bahwa A terhubung dengan B dan C.
- Baris B menunjukkan bahwa B terhubung dengan C.
- Baris C menunjukkan bahwa C terhubung dengan D.
- Baris D menunjukkan bahwa D tidak terhubung dengan node manapun.

## Graph

Graph (graf) adalah struktur data yang digunakan untuk merepresentasikan hubungan antara objek. Dalam istilah sederhana, graf adalah kumpulan simpul (nodes) dan sisi (edges) yang menghubungkan pasangan simpul. Struktur ini sangat berguna dalam berbagai bidang, termasuk ilmu komputer, matematika, dan pemrosesan bahasa alami.

**Komponen Utama Graf**

1. Simpul (Node): 
   Simpul, juga dikenal sebagai vertex, adalah elemen dasar dalam graf. Setiap simpul mewakili entitas atau objek tertentu. Dalam konteks pemrosesan bahasa alami, simpul bisa mewakili kalimat, kata, atau dokumen.

2. Sisi (Edge): 
   Sisi adalah hubungan antara dua simpul. Sisi dapat bersifat:
   - Arah (Directed): Menunjukkan bahwa hubungan dari satu simpul ke simpul lainnya memiliki arah tertentu. Misalnya, dalam graf yang menggambarkan hubungan antara kalimat, jika kalimat A mengarah ke kalimat B, ini menunjukkan bahwa A memberikan informasi yang relevan untuk B.
   - Tidak Arah (Undirected): Menunjukkan bahwa hubungan antara dua simpul tidak memiliki arah tertentu. Contohnya, jika dua kalimat saling berhubungan, kita dapat menggambarkannya dengan sisi yang tidak terarah.

**Tipe-Tipe Graf**

1. Graf Berarah (Directed Graph): 
   Dalam graf berarah, setiap sisi memiliki arah, mengindikasikan hubungan satu arah antara dua simpul. Contohnya, graf yang menunjukkan hubungan antar situs web, di mana satu situs mengarah ke situs lainnya.

2. Graf Tidak Berarah (Undirected Graph): 
   Dalam graf tidak berarah, hubungan antara simpul bersifat dua arah. Contohnya, graf sosial di mana simpul mewakili orang dan sisi mewakili pertemanan antara mereka.

3. Graf Berat (Weighted Graph): 
   Dalam graf berat, setiap sisi memiliki bobot yang menunjukkan kekuatan atau jarak dari hubungan tersebut. Misalnya, dalam graf rute perjalanan, bobot bisa mewakili jarak atau waktu tempuh antara dua titik.

4. Graf Tidak Berat (Unweighted Graph): 
   Dalam graf tidak berat, semua sisi dianggap setara dan tidak memiliki bobot.

**Representasi Graf**

Graf dapat direpresentasikan dalam beberapa cara, di antaranya:

1. Matriks Adjacency: 
   Matriks persegi di mana elemen di baris dan kolom merepresentasikan simpul, dan nilai dalam matriks menunjukkan ada tidaknya hubungan antara simpul-simpul tersebut. Misalnya, jika ada sisi antara simpul A dan B, maka elemen matriks pada posisi (A, B) akan diisi dengan 1 (atau bobot) dan 0 jika tidak ada sisi.

2. Daftar Ketetanggaan (Adjacency List): 
   Dalam representasi ini, setiap simpul memiliki daftar yang berisi semua simpul yang terhubung langsung dengan simpul tersebut. Ini lebih efisien dalam penggunaan memori, terutama untuk graf yang jarang terhubung (sparse).

## Closeness Centrality

**Closeness centrality** adalah salah satu ukuran yang digunakan dalam teori graf untuk menentukan seberapa dekat sebuah node (simpul) dengan node lainnya dalam sebuah graf. Konsep ini berguna untuk mengidentifikasi simpul yang memiliki posisi strategis dalam hal keterhubungan dan pengaruh dalam jaringan. Dalam konteks pemrosesan bahasa alami, closeness centrality dapat digunakan untuk menentukan kalimat atau kata yang paling relevan atau penting dalam sebuah dokumen.

**Konsep Dasar Closeness Centrality**

1. **Definisi**:
   Closeness centrality mengukur seberapa cepat sebuah simpul dapat menjangkau simpul lainnya dalam graf. Semakin dekat sebuah simpul dengan semua simpul lainnya, semakin tinggi nilai closeness centrality-nya. 

2. **Rumus**:
   Nilai closeness centrality untuk simpul $v$ dalam graf dapat dihitung dengan rumus:

   $$
   C(v) = \frac{1}{\sum_{u \in V} d(v, u)}
   $$

   Di mana:
   - $C(v)$ adalah nilai closeness centrality untuk simpul $v$.
   - $V$ adalah himpunan semua simpul dalam graf.
   - $d(v, u)$ adalah jarak terpendek antara simpul $v$ dan simpul $u$.

   Dalam beberapa konteks, rumus ini dapat dimodifikasi untuk mengalikan dengan $(|V| - 1)$ sehingga nilai centrality berkisar antara 0 dan 1, di mana $ |V| $ adalah jumlah total simpul dalam graf.

3. **Interpretasi**:
   - **Nilai Tinggi**: Simpul dengan nilai closeness centrality tinggi dapat menjangkau semua simpul lain dengan cepat, menunjukkan bahwa simpul tersebut memiliki posisi strategis dalam jaringan.
   - **Nilai Rendah**: Sebaliknya, simpul dengan nilai rendah menunjukkan bahwa simpul tersebut lebih terisolasi dan membutuhkan lebih banyak langkah untuk menjangkau simpul lain.

## Summarization

**Summarization** (penyusunan ringkasan) adalah proses dalam pemrosesan bahasa alami (Natural Language Processing, NLP) yang bertujuan untuk menghasilkan versi ringkas dari sebuah dokumen atau teks yang mempertahankan informasi penting dan inti dari teks aslinya. Summarization dapat digunakan dalam berbagai konteks, seperti menyederhanakan dokumen panjang, membantu pembaca mendapatkan gambaran umum tentang konten, dan meningkatkan efisiensi dalam pengolahan informasi.

**Tipe-Tipe Summarization**

1. **Extractive Summarization**:
   - Pada metode ini, ringkasan dibuat dengan memilih dan menggabungkan kalimat-kalimat penting langsung dari teks asli. 
   - Proses ini melibatkan teknik untuk menilai relevansi kalimat, seperti menggunakan nilai frekuensi kata, model berbasis graf (misalnya, PageRank), dan teknik pembelajaran mesin.
   - Kelebihan dari metode ini adalah kesederhanaan dan kemudahan pemahaman, karena kalimat yang dipilih adalah bagian asli dari teks.
   - Contoh: Menggunakan algoritma TF-IDF untuk menilai pentingnya kalimat dalam teks dan kemudian memilih kalimat dengan nilai tertinggi untuk dimasukkan ke dalam ringkasan.

2. **Abstractive Summarization**:
   - Dalam pendekatan ini, ringkasan dihasilkan dengan menghasilkan kalimat baru yang mungkin tidak ada dalam teks asli. Proses ini lebih mirip dengan cara manusia merangkum informasi.
   - Teknik ini biasanya melibatkan model pembelajaran mendalam (deep learning), seperti jaringan saraf berulang (Recurrent Neural Networks, RNN) dan Transformer.
   - Kelebihan dari metode ini adalah fleksibilitasnya untuk menghasilkan ringkasan yang lebih alami dan terfokus, meskipun tantangannya adalah menjaga keakuratan dan kesesuaian informasi.
   - Contoh: Menggunakan model seperti GPT (Generative Pre-trained Transformer) untuk menghasilkan ringkasan dari teks panjang.

**Langkah-Langkah dalam Proses Summarization**

Berikut adalah penjelasan rinci untuk setiap langkah dalam proses summarization menggunakan graph:

 1. **Pengolahan Dokumen**
Pengolahan dokumen adalah langkah awal yang krusial dalam proses pemrosesan bahasa alami. Dalam tahap ini, teks dokumen yang akan dianalisis mengalami beberapa transformasi untuk meningkatkan kualitas analisis selanjutnya. Pertama, seluruh teks diubah menjadi huruf kecil (lowercase) agar tidak ada perbedaan dalam penulisan yang memengaruhi hasil analisis. Selanjutnya, tanda baca dihapus dari teks untuk menghindari gangguan dalam pengolahan. Setelah itu, kata-kata yang dianggap tidak penting atau umum, yang dikenal sebagai stopwords, dibuang dari teks. Akhirnya, dokumen dipecah menjadi kalimat-kalimat terpisah (tokenisasi per kalimat), yang akan menjadi unit dasar dalam analisis. Pendekatan ini memastikan bahwa fokus analisis berada pada kalimat, yang memiliki makna lebih dalam konteks keseluruhan dokumen.

 2. **Ekstraksi Fitur Menggunakan TF-IDF**
Setelah pengolahan dokumen selesai, langkah berikutnya adalah ekstraksi fitur menggunakan metode TF-IDF (Term Frequency-Inverse Document Frequency). TF-IDF adalah teknik yang digunakan untuk menghitung bobot atau kepentingan dari setiap kalimat dalam dokumen berdasarkan frekuensi kemunculan kata dalam kalimat tersebut dan seberapa umum kata tersebut di seluruh dokumen. Dengan menggunakan TF-IDF, kalimat-kalimat yang memiliki kata-kata unik dan informatif akan mendapatkan nilai yang lebih tinggi, sementara kalimat dengan kata-kata yang umum akan memiliki bobot yang lebih rendah. Proses ini sangat penting karena nilai TF-IDF yang dihasilkan akan digunakan untuk menentukan relevansi kalimat-kalimat dalam dokumen.

 3. **Menghitung Cosine Similarity**
Setelah mendapatkan matriks TF-IDF, langkah selanjutnya adalah menghitung cosine similarity antar kalimat. Cosine similarity adalah ukuran yang digunakan untuk mengukur seberapa mirip dua vektor (dalam hal ini, kalimat yang direpresentasikan sebagai vektor TF-IDF) berdasarkan sudut di antara mereka dalam ruang vektor. Hasil dari perhitungan ini adalah matriks cosine similarity, di mana setiap elemen mencerminkan tingkat kesamaan antara dua kalimat. Matriks ini sangat berguna untuk memahami hubungan antar kalimat dan menentukan kalimat mana yang memiliki konten yang mirip atau berhubungan erat.

 4. **Matriks Adjacency**
Setelah menghitung cosine similarity, langkah berikutnya adalah mengonversi matriks cosine similarity menjadi matriks adjacency. Matriks adjacency adalah representasi grafis dari hubungan antar kalimat, di mana setiap elemen menunjukkan apakah ada hubungan yang cukup kuat antara dua kalimat. Untuk menentukan hubungan yang cukup kuat, diterapkan threshold tertentu, misalnya 0.5. Jika nilai cosine similarity antara dua kalimat di atas threshold tersebut, maka hubungan antar kalimat tersebut dianggap ada (dinyatakan dengan nilai 1) dan sebaliknya (dinyatakan dengan nilai 0). Dengan cara ini, matriks adjacency menyederhanakan hubungan yang relevan antara kalimat-kalimat dalam dokumen.

 5. **Membangun Graph**
Dari matriks adjacency yang telah dibentuk, langkah selanjutnya adalah membangun graph yang merepresentasikan hubungan antar kalimat. Dalam graph ini, setiap node mewakili kalimat, sedangkan setiap edge menggambarkan hubungan antar kalimat berdasarkan nilai dari matriks adjacency. Graph ini memberikan cara visual untuk memahami struktur hubungan dalam dokumen dan membantu dalam analisis lebih lanjut. Penggunaan graph dalam NLP memungkinkan pemodelan hubungan kompleks antar kalimat dengan cara yang lebih intuitif dan terstruktur.

 6. **Menghitung Closeness Centrality**
Setelah graph dibangun, langkah selanjutnya adalah menghitung closeness centrality untuk setiap node dalam graph. Closeness centrality adalah ukuran yang menunjukkan seberapa dekat sebuah node dengan semua node lain di dalam graph. Semakin tinggi nilai closeness centrality dari sebuah kalimat, semakin penting posisi kalimat tersebut dalam konteks keseluruhan dokumen. Dengan menghitung closeness centrality, kita dapat mengidentifikasi kalimat-kalimat yang paling representatif dan berkontribusi besar terhadap pemahaman dokumen secara keseluruhan.

 7. **Ranking Node**
Setelah menghitung nilai closeness centrality, langkah berikutnya adalah meranking kalimat berdasarkan nilai tersebut. Dalam proses ini, kalimat-kalimat diurutkan berdasarkan nilai centrality mereka, dan disni diambil contoh tiga kalimat dengan nilai tertinggi dipilih sebagai hasil ringkasan. Kalimat-kalimat ini dianggap paling informatif dan relevan, mewakili inti dari isi dokumen. Proses ranking ini membantu dalam mengidentifikasi konten penting yang perlu disertakan dalam ringkasan.

 8. **Summarization**
Langkah terakhir dalam proses ini adalah menghasilkan ringkasan dari dokumen menggunakan kalimat-kalimat teratas yang telah dipilih. Kalimat-kalimat ini disusun kembali untuk membentuk ringkasan yang koheren dan informatif. Dengan pendekatan ini, ringkasan yang dihasilkan diharapkan dapat mencerminkan inti dari dokumen secara akurat, memberikan pemahaman yang lebih cepat dan efisien bagi pembaca. Proses summarization ini sangat bermanfaat dalam berbagai aplikasi, seperti menyajikan informasi penting dari laporan, artikel, atau dokumen panjang lainnya dengan cara yang ringkas dan jelas.

```python
# Tokenisasi teks
berita_tokenized = sent_tokenize(berita_cleaned)
berita_tokenized
```

### Stopword Removal

```python
# Hapus stopwords
stop_words = set(stopwords.words('indonesian'))

custom_stopwords = {'kompas.com'}
stop_words.update(custom_stopwords)
```

```python
print(stop_words)
```

```python
# Proses setiap kalimat, menghapus stopwords
berita_no_stopwords = []
for sentence in berita_tokenized:
    # Tokenisasi kata di dalam setiap kalimat
    words = sentence.split()
    # Hapus stopwords
    words_no_stop = [word for word in words if word not in stop_words]
    # Gabungkan kembali kata-kata tanpa stopwords menjadi kalimat
    cleaned_sentence = ' '.join(words_no_stop)
    berita_no_stopwords.append(cleaned_sentence)

berita_no_stopwords
```

## TF-IDF

```python
vect = TfidfVectorizer()
vect_matrix = vect.fit_transform(berita_no_stopwords)
```

```python
terms = vect.get_feature_names_out()
tfidf = pd.DataFrame(data=vect_matrix.toarray(), columns=terms)
tfidf
```

## Cosine Similarity

```python
# Menghitung cosine similarity antara semua pasangan kalimat
cosine = cosine_similarity(vect_matrix, vect_matrix)

# Membuat DataFrame untuk cosine similarity
cosine_sim = pd.DataFrame(cosine, columns=range(len(berita_no_stopwords)), index=range(len(berita_no_stopwords)))

# Menampilkan matriks cosine similarity
cosine_sim
```

## Matrik Adjensi

```python
# Mengatur threshold untuk membangun matriks adjensi
threshold = 0.2

# Menangani nilai NaN pada cosine_sim
cosine_sim_filled = cosine_sim.fillna(0)

# Membuat matriks adjensi berdasarkan cosine similarity
adjacency = (cosine_sim_filled.values > threshold).astype(int)

adjacency_matrix = pd.DataFrame(adjacency, 
                             columns=[f'{i}' for i in range(len(berita_no_stopwords))],
                             index=[f'{i}' for i in range(len(berita_no_stopwords))])

adjacency_matrix
```

### Graph

```python
# Membuat graf
G = nx.Graph()

# Tambahkan node (kalimat)
for i in range(len(adjacency_matrix)):
    G.add_node(i)

# Tambahkan edge (hubungan) antara kalimat berdasarkan matriks adjensi
for i in range(len(adjacency_matrix)):
    for j in range(len(adjacency_matrix)):
        similarity = adjacency_matrix.iloc[i, j]  
        if similarity > 0 and i != j:  
            G.add_edge(i, j, weight=similarity)

# Visualisasikan grafik dengan tata letak "circular"
pos = nx.circular_layout(G)

# Membuat label node yang hanya menggunakan nomor indeks
labels = {i: f'K {i}' for i in G.nodes()}

plt.figure(figsize=(6, 6))
nx.draw(G, pos, with_labels=True, labels=labels, node_size=900, node_color='skyblue', edge_color='gray')
plt.title('Graf Hubungan Kalimat Berdasarkan Matriks Adjacency')
plt.show()
```
